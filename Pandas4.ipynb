# Pivot Table

```python
df = pd.read_csv('profit1.csv')
df
```

```python
df1 = pd.pivot_table(df,index=['Catg'],aggfunc='mean')
df1
```

```python
df.groupby(by='Catg').mean()
```

```python
df1 = pd.pivot_table(df,index=['Catg'],columns='HQ',aggfunc='mean')
df1
```

```python
df.groupby(by=['Catg','HQ']).mean()
```

```python
df1 = pd.pivot_table(df,index=['Catg'],columns='HQ',aggfunc='max')
df1
```

```python
df1 = pd.pivot_table(df,index=['Catg'],columns='HQ',aggfunc=['max','min'])
df1
```

```python
df1.index
```

```python
df1.columns
```

```python
df1 = pd.pivot_table(df,index=['Catg'],columns='HQ',values=['Vol','Rev'],aggfunc=['max','min'])
df1
```

```python
df1 = pd.pivot_table(df,index=['Catg','Sector'],columns='HQ',values=['Vol','Rev'],aggfunc=['max','min'])
df1
```

```python
df1 = pd.pivot_table(df,index=['Catg','Sector'],columns='HQ',values=['Vol','Rev'],aggfunc=['max','min'],fill_value=0)
df1
```

# Merge

```python
dfleft = pd.read_csv('profit11.csv')
dfright = pd.read_csv('profit12.csv')
dfm = pd.merge(dfleft,dfright,on='HQ',how='outer')
dfm
```

```python
dfm = pd.merge(dfleft,dfright,on='HQ',how='inner',indicator=True)
dfm
```

```python
dfm = pd.merge(dfleft,dfright,on='HQ',how='left',indicator=True)
dfm
```

```python
dfm = pd.merge(dfleft,dfright,on='HQ',how='right',indicator=True)
dfm
```
