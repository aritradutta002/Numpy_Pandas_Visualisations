```python
col_labels
```

# Mapping

```python
df1 = pd.read_csv('profit1.csv')
df1
```

```python
df1['Catg'].value_counts()
```

```python
df1['Catg'].count()
```

```python
city2zone = {'NCR':'NZ','Mum':'WZ','Kol':'EZ','Che':'SZ','Blr':'SZ'}
```

```python
df1['Zone'] = df1['HQ'].map(city2zone)
```

```python
df1
```

```python
def vsales_grp(v):
    if v > 950:
        return 'Good Sales'
    else:
        return 'Average Sales'
```

```python
df1['Vol_Grp'] = df1['Vol'].map(vsales_grp)   # function as parameter
df1
```

```python
def com(e):
    if e > 650:
        return 'Liberal'
    else:
        return 'Controlled'
```

```python
df1['Company'] = df1['Exp'].map(com)
df1
```
