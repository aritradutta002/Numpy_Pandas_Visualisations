### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 
### Assign it to a variable called users and use the 'user_id' as index

```python
user_id = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',sep='|')
user_id
```

### Step 3. See the last 10 entries

```python
user_id.tail(10)
```

### Step 4. How many different occupations are in this dataset?

```python
user_id[user_id.duplicated('occupation') == False]['occupation'].count()
```

### Step 5. What is the most frequent occupation?

```python
user_id.groupby(by='occupation').size().sort_values(ascending=False)
```

```python
pd.pivot_table(user_id,index='occupation',aggfunc='size').sort_values(ascending=False)
```

### Step 6. Write a code to show the average age for all individual 'occupation'

```python
user_id.groupby(by='occupation')['age'].mean()
```

```python
user_id.pivot_table(index='occupation',values='age',aggfunc='mean')
```

### Step 7. Write a code to show the mean for 'Age' and mode for 'Occupation' on the basis of gender.

```python
user_id.pivot_table(index='gender',values=['age','occupation'] , aggfunc=({'age':'mean','occupation':'mode'}))
```

### Step 8. Write a code to show the average age of all the genders as per there respective occupation using pivot_table

```python
user_id.pivot_table(index='occupation',columns='gender',aggfunc={'age':'mean'})
```

### Step 9. Map all the M to male and F to female in gender: (Use map)

```python
gendermap = {'M':'male','F':'female'}
user_id['gender'] = user_id['gender'].map(gendermap)
user_id
```

### Step 10. Map all the M to male and F to female in gender: (Use replace)

```python
user_id.replace(gendermap)
```

### Step 11. Write a code to RENAME the column from gender to sex and assign it to the new dataframe named as df1

```python
df1 = user_id.rename(mapper={'gender':'sex'},axis=1)
df1
```

### Step 12 Write a code to drop the column 'Zipcode'  from df1 and assign it to the new dataframe named as df2.

```python
df2 = df1.drop(labels='zip_code',axis=1)
df2
```

### Step 13 Convert all the occupation that start with a vowel to upper using lambda and apply.

```python
df2['occupation'].apply(lambda x: x[0].upper()+x[1:] if x.startswith('a'or'e'or'i'or'o'or'u') else x)
```
