```python
import numpy as np
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])

#output: 1,3,5,6,8,10

#set: union|, intersection&, setdiff, exor== symettric difference^
```

```python
(set(sr1) - set(sr2) ) | (set(sr2)- set(sr1) )
```

```python
set(sr1) ^ set(sr2)
```

```python
a=np.setxor1d(sr1,sr2)   #1,2,3,4,5 ^ # 2,4,6,8,10
a
# 01= 1
# 10= 1
# 11=0
# 00=0
```

```python
set(sr1) | set(sr2)   #01 , 10, 11, 00
```

```python
u = np.union1d(sr1,sr2)
print(u)
u1 = np.intersect1d(sr1,sr2)
print(u1)
final = np.setdiff1d(u,u1)
print(final)
```

The default manner to create a DataFrame from python is to use a list of dictionaries. In this case each dictionary key is used for the column headings. A default index will be created automatically:

```python
GL = ({"Location":"Bangalore","Jan":150,"Feb":80},
     {"Location":"Chennai","Jan":200,"Feb":70})
pd.DataFrame(GL)
```

<span style="font-family: Arial; font-weight:bold;font-size:1.50em;color:#7604ce;"> DataFrame using List
    
The other option for creating your DataFrames from python is to include the data in a list structure.

The first approach is to use a row oriented approach using pandas from_records . This approach is similar to the dictionary approach but you need to explicitly call out the column labels.

```python
Image('8.png')
```

```python
GL = [("Bangalore",150,80),
      ("Chennai",200,70)]
Labels = ["Location","Jan","Feb"]
df = pd.DataFrame.from_records(GL,columns=Labels)
df
```

```python
GL = [("Location", ["Bangalore","Chennai"]),
       ("Jan",[150,200]),
       ("Feb",[80,70])]

df = pd.DataFrame.from_items(GL)
df
```

```python
from collections import OrderedDict 
GL = [("Location", ["Bangalore","Chennai"]),
       ("Jan",[150,200]),
       ("Feb",[80,70])]
print(OrderedDict(GL))
df = pd.DataFrame.from_dict(OrderedDict(GL))
df
```

```python
#1. Create a Dataframe with Name, Subjects and Marks for the Students
#Create a DataFrame
d = {
'Name':['Alisa','Bobby','Cathrine','Katry','Alisa','Bobby','Cathrine','Katry',
'Alisa','Bobby','Cathrine','Katry','Alisa','Bobby','Cathrine','Katry'],
'Subject':['Mathematics','Mathematics','Mathematics','Mathematics','Science','Science','Science','Science',
'History','History','History','History','Economics','Economics','Economics','Economics'],
'Score':[62,47,55,78,74,31,77,89,85,63,42,57,62,89,85,91]}
 
data = pd.DataFrame(d,columns=['Name','Subject','Score'])
data
```

```python
# Rank the dataframe by ascending and descending order   #SCORE
```

```python
data['Rank'] = data['Score'].rank(ascending = False)
```

```python
data.sort_values(by='Score')
```

```python
data['Rank'] = data['Score'].rank(ascending = True)
data
```

```python
# Dense rank
```

```python
data['Rank']=data['Score'].rank(method='dense',ascending=True)
```

```python
#Rank the dataframe by Maximum rank if found 2 values are same
```

```python
data['Rank'] = data['Score'].rank(method='max',ascending=False)
data.sort_values(by='Rank')

```

```python
data['Rank'] = data['Score'].rank(method='min',ascending=False)
data.sort_values(by='Rank')
```

```python
data.groupby(by='Subject')['Score'].sum().rank(ascending=False,method='dense')
```

```python
data['group rank']=data.groupby(by='Subject')['Score'].rank(ascending=False,method='dense')
data
```

```python
data.groupby(by='Subject').get_group('History')
```

```python
data.groupby(by='Subject').rank(ascending=0)

```
